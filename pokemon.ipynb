```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
pokemon = pd.read_csv('../Datasets/pokemon.csv')
pokemon
```

```python
X = pokemon[['total','hp','attack','defense','sp_attack','sp_defense','speed']]
y = pokemon.legendary

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
```

```python
model = RandomForestClassifier(max_leaf_nodes=10)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print("Accuracy Score:",accuracy_score(predictions, y_test))
```

```python
normal = [[490,75,110,105,30,70,100]]
legen = [[590,99,83,91,125,83,109]]

print("Normal Pokemon",model.predict(normal))
print("Legendary Pokemon",model.predict(legen))
```
